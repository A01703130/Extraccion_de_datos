```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Convertir en dataframe
df= pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
df
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
#Imprimo los primeros 5 registro del dataframe
df.head()
```

# **FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Se filtra dentro de Nombre_cliente, público en general
filtro1=df[df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL"]
filtro1
```

```python
filtro1.to_csv("Filtro_por_objeto.csv")
#descargar archivo filtrado en csv
from google.colab import files   
files.download("Filtro_por_objeto.csv")
```

```python
#Filtro por filas
#Se están filtrando las filas 6,20 y 100
filtro2=df.iloc[ [6, 20, 100] , : ]  #Filas no consecutivas 
filtro2
```

```python
filtro2.to_csv("filtro_por_filas.csv")
from google.colab import files   
files.download("filtro_por_filas.csv")
```

```python
#Filtro por columnas
#Se están filtrando las primeras 4 columnas
filtro3=df.iloc[ : , 1:4]       #Columnas consecutivas 
filtro3
```

```python
filtro3.to_csv("filtro_por_columnas.csv")
from google.colab import files   
files.download("filtro_por_columnas.csv")
```

```python
#Filtro de filas vs columnas 
#Se filtran filas y columnas por nombre del cliente y fecha que se indica posteriormente
df1= pd.read_excel("Detalle precios y productos fabricados 2022.xlsx", index_col=1)
df1
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas 
#Aqui se observa como filtre por la fechas del documento y por el nombre del cliente 
filtro4=df1.loc[["2022-01-05","2022-12-23"], ["NOMBRE_CLIENTE"]]
filtro4
```

```python
filtro4.to_csv("filtro_filasvs_columnas.csv")
#descargar archivo filtrado en csv
from google.colab import files   
files.download("filtro_filasvs_columnas.csv")
```

```python
#Se muestran los primeros 10 registros
filtro5=df.head(10)
filtro5
```

```python
filtro5.to_csv("filtro_por_cabecera.csv")
from google.colab import files   
files.download("filtro_por_cabecera.csv")
```

```python
#Filtro por comparacion
#Aqui se filtran  los costos que mayores a 100 
filtro6=df[df["COSTO_UNITARIO_CALCULADO"] > 100]
filtro6
```

```python
filtro6.to_csv("filtro_por_comparacion.csv")
from google.colab import files   
files.download("filtro_por_comparacion.csv")
```

```python
#Filtro Y
#Aquí se filtran los costos unitarios  mayores a 100 y la fecha igual a 1 de mayo del 2022
filtro7=df[(df["COSTO_UNITARIO_CALCULADO"] < 100) & (df["FECHA_DOC"] == "2022-01-05")]
filtro7
```

```python
filtro7.to_csv("filtro_y.csv")
from google.colab import files   
files.download("filtro_y.csv")
```

```python
#Filtro o
#Aquí se filtran los costos unitarios mayores a 100 o la fecha igual a 1 de mayo del 2022
filtro8=df[(df["COSTO_UNITARIO_CALCULADO"] < 100) | (df["FECHA_DOC"] == "2022-01-05")]
filtro8
```

```python
filtro8.to_csv("filtro_O.csv")
from google.colab import files   
files.download("filtro_O.csv")
```

```python
#Filtro not
#Aquí se filtran los costos unitarios menores o iguales  a 100 no la fecha igual a 1 de mayo del 2022
filtro9=df[~(df["COSTO_UNITARIO_CALCULADO"] >= 100) & ~(df["FECHA_DOC"] == "2022-01-05")]
filtro9
```

```python
filtro9.to_csv("filtro_not.csv")
from google.colab import files   
files.download("filtro_not.csv")
```
